```python
import torch 
```

```python
random_int = torch.rand(5)

print(random_int)

print(random_int.ndim)   
```

```python
r_n = torch.rand(2,5,3)

print(r_n)

print(r_n.ndim)
print(r_n.shape)
```

```python
torch.manual_seed(0)  # For reproducibility

r_t = torch.rand(2, 5, 3)
print(r_t)
print(r_t.ndim)
print(r_t.shape)    
```

```python
import matplotlib.pyplot as plt 

uniform_noise = torch.rand((256,256))

plt.imshow(uniform_noise,cmap= 'Greys')
plt.title('Uniform Noise')
plt.axis('off') 
plt.show()
```

```python
#Rgb image 
import matplotlib.pyplot as plt 

uniform_noise_rgb = torch.rand((256,256,3))

plt.imshow(uniform_noise_rgb)
plt.title('Uniform Noise')
plt.axis('off') 
plt.show()
```
